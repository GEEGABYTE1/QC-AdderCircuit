```python
from qiskit import QuantumCircuit
from qiskit.providers.aer import AerSimulator
import random
sim = AerSimulator()
```

```python
def int_to_bin(integer):
        exponents = []
        counter = 0
        remainder = 0
        mod = 0
        while integer >= 1:
            while mod != 1:
                result_mod = integer // (2 ** counter)
                mod = result_mod
                counter += 1
            counter -= 1
            exponents.append(counter)
            integer = integer - (2 ** counter)
            counter = 0 
            mod = 0
        
        binary_num = ''
        for i in range(max(exponents), -1, -1):
            if i in exponents:
                binary_num += '1 '
            else:
                binary_num += '0 '
        
        return binary_num
```

```python
try:
    user_input = int(input('Please input your number: '))
except:
    print('Input not valid') 
bin_num = int_to_bin(user_input)

```

```python
size_of_circuit = int(input(('Size of Circuit: '))) 
qc = QuantumCircuit(size_of_circuit, 2)
for string in bin_num:
    if string == ' ':
        continue
    else:
        qc.x(int(string))

qubits = []
counter = 0
qubits = []

for i in range(size_of_circuit - 1):
    if i == 0:
        qc.cx(0, 1)
        qubits.append((0, 1))
    else:
        qc.cx(i, i - 1)
        qubits.append((i, i - 1))

qc.ccx(0, size_of_circuit // 2, size_of_circuit - 1)

print(qubits)       # (3, 2) Qubit Check Fix




    


    


```

```python

qc.measure(0, 1)
qc.measure(1, 0)
qc.measure(2, 1)
job = sim.run(qc)
result = job.result()
result.get_counts()
qc.draw()


```
